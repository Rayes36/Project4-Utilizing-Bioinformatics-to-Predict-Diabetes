```python
# Loading Complementary DNA Sequence into an alignable file
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
DNAseq = SeqIO.read('/content/drive/MyDrive/bio/data/MN908947.fna', "fasta")
```

```python
# Loading sequence
sequence = '/content/drive/MyDrive/bio/data/MN908947.txt'
with open(sequence) as text:
    print (text.read(100000))
```

```python
#Obtaining DNA
DNA = DNAseq.seq
print(DNA)
```

```python
#Obtain mRNA Sequence
mRNA = DNA.transcribe()
print(mRNA)
```

```python
# Obtain Amino Acid Sequence from mRNA
Amino_Acid = mRNA.translate()
print(Amino_Acid)
print("Length of Protein:",len(Amino_Acid))
print("Length of Original mRNA:",len(mRNA))
```

```python
#Create a codon table
from Bio.Data import CodonTable
print(CodonTable.unambiguous_rna_by_name['Standard'])
```

```python
#Identify and detecting all the Proteins (chains of amino acids)
Proteins = Amino_Acid.split('*')
Proteins
```

```python
#Remove chains smaller than 20 amino acids long (cleaning data)
for i in Proteins[:]:
    if len(i) < 20:
        Proteins.remove(i)
```

```python
Proteins
```

```python
# NCBOWWW -> world wide database for protein
from Bio.Blast import NCBIWWW
result_handle = NCBIWWW.qblast("blastp", "nt", Proteins)
```

```python
import io
content = result_handle.read()

print(content)
```

```python
import xml.etree.ElementTree as ET

# Parse the XML data
root = ET.fromstring(content)

# Extract relevant information
program = root.find('BlastOutput_program').text
version = root.find('BlastOutput_version').text
reference = root.find('BlastOutput_reference').text
database = root.find('BlastOutput_db').text
query_id = root.find('BlastOutput_query-ID').text
query_def = root.find('BlastOutput_query-def').text
query_len = root.find('BlastOutput_query-len').text

# Print the extracted information
print("Program:", program)
print("Version:", version)
print("Reference:", reference)
print("Database:", database)
print("Query ID:", query_id)
print("Query Definition:", query_def)
print("Query Length:", query_len)

# Extract additional information from Iteration_stat
statistics = root.find('.//Iteration_stat/Statistics')
db_num = statistics.find('Statistics_db-num').text
db_len = statistics.find('Statistics_db-len').text
hsp_len = statistics.find('Statistics_hsp-len').text
eff_space = statistics.find('Statistics_eff-space').text
kappa = statistics.find('Statistics_kappa').text
lambda_val = statistics.find('Statistics_lambda').text
entropy = statistics.find('Statistics_entropy').text

# Print additional extracted information
print("Database Number:", db_num)
print("Database Length:", db_len)
print("HSP Length:", hsp_len)
print("Effective Space:", eff_space)
print("Kappa:", kappa)
print("Lambda:", lambda_val)
print("Entropy:", entropy)
```

```python
from Bio.Blast import NCBIXML
from Bio import ExPASy
from Bio import SwissProt
result_handle_fin = io.StringIO(content)
blast_record = NCBIXML.read(result_handle_fin)
if blast_record.alignments:

  top_hit = blast_record.alignments[0]
  accession = top_hit.accession
  protein_name = top_hit.hit_def


  handle = ExPASy.get_sprot_raw(accession)
  record = SwissProt.read(handle)


  gene_name = record.gene_name
  protein_names = record.data['GN']
  protein_functions = record.comments

  print("Accession:", accession)
  print("Protein Name:", protein_name)
  print("Gene Name:", gene_name)
  print("Protein Names:", protein_names)
  print("Protein Functions:")
  for function in protein_functions:
      print(function)
else:
    print("No alignments found in the BLAST result.")
```

```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
MW = []
aromaticity =[]
AA_Freq = []
IsoElectric = []
for j in Proteins[:]:
    a = ProteinAnalysis(str(j))
    MW.append(a.molecular_weight())
    aromaticity.append(a.aromaticity())
    AA_Freq.append(a.count_amino_acids())
    IsoElectric.append(a.isoelectric_point())
```

```python
MW = pd.DataFrame(data = MW,columns = ["Molecular Weights"] )
MW.head()
```

```python
# Plot Molecular Weights to see the differences of Covid-19 Protein
sns.set_style('whitegrid');
plt.figure(figsize=(10,6));
sns.histplot(MW,kde=False);
plt.title("SARS-CoV-2 Protein Molecular Weights Distribution");
```
